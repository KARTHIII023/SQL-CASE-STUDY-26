Task 8
Merge all four datasets into one DataFrame.


```python
merged = pd.merge(pop, gdp, on="country", how="inner")
```

```python
merged = pd.merge(merged, user, on="country", how="inner")
```

```python
merged = pd.merge(merged, lit, on="country", how="inner")
```

```python
merged
```

Task 9
Create the following calculated columns:
1. Internet Penetration (%)
2. GDP Per Capita


```python
merged["Internet_Penetration"] = (merged["internet_users"] / merged["population"]) * 100
merged["Internet_Penetration"]
```

```python
merged["gdp_per"] = merged["GDP"] / merged["population"]
merged["gdp_per"] = merged["gdp_per"].round(2)
merged
```

Part D – Load Data
Task 10
Create a SQLite database. Store the final merged dataset into a table named
‘country_statistics’

```python
import sqlite3
```

```python
conn = sqlite3.connect("country_statistics.db")
```

```python
merged.to_sql(
    "country_statistics",
    conn,
    if_exists="replace",
    index=False
)
```

12. Find the average literacy rate.

```python
query2 = """
SELECT AVG(Literacy_Rate)
FROM country_statistics
"""

cursor = conn.cursor()
cursor.execute(query2)
result = cursor.fetchall()
result
```

13. Display all countries having literacy rate above 90%.

```python
query3 = """
SELECT country, Literacy_Rate
FROM country_statistics
WHERE Literacy_Rate > 90
"""

cursor = conn.cursor()
cursor.execute(query3)
result = cursor.fetchall()
result
```

14. Find the Top 5 countries with the highest internet penetration.

```python
query4 = """
SELECT country, Internet_Penetration
FROM country_statistics
ORDER BY Internet_Penetration DESC
LIMIT 5
"""

cursor = conn.cursor()
cursor.execute(query4)
result = cursor.fetchall()
result
```

15. Calculate the total population.

16. Find the country with the highest GDP per capita.

```python
query6 = """
SELECT country, gdp_per
FROM country_statistics
ORDER BY gdp_per DESC
LIMIT 1
"""

cursor = conn.cursor()
cursor.execute(query6)
result = cursor.fetchall()
result
```
